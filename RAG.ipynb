```python
#%%writefile Expert_Nutritionist_App.py
import streamlit as st
from PIL import Image
import google.generativeai as genai

# Default prompt
default_prompt = """As a nutrition expert with an image-based analysis focus, your task is to examine food items in an image and determine their total calories. Provide a detailed breakdown of each food item's calorie content in the following format:

Item 1 - Number of calories
Item 2 - Number of calories
...
Following this, assess the overall healthiness of the meal. Include a percentage split for carbohydrates, proteins, fats, sugar, and total calories in the meal. Finally, offer suggestions for improving the meal's health status by recommending items for removal and addition, particularly if the current composition is deemed unhealthy.
And also give in below formate where to improve in daily activities, such as exercise routines, calorie intake, and water consumption.
"""

# Initialize the Streamlit app
st.set_page_config(
    page_title="Sai Kundan Suddapalli",
    layout="wide",
    initial_sidebar_state="expanded"
)

# Add custom icons using HTML and CSS
st.markdown("""
    <style>
        .icon {
            display: flex;
            align-items: center;
            padding: 10px;
        }
        .icon img {
            margin-right: 5px;
        }
    </style>
""", unsafe_allow_html=True)


# Two functions for API communication
model = genai.GenerativeModel('gemini-pro-vision')
genai.configure(api_key='AIzaSyBXm5Ty_irW06QR3MN3BAd7ejhqYCbyEFE')

def say(input, image, user_prompt):
    try:
        response = model.generate_content([input, image[0], user_prompt])
        return response.text
    except Exception as e:
        # Handle the error gracefully
        return f"An error occurred: {str(e)}"

def input_image_details(uploaded_file):
    if uploaded_file is not None:
        bytes_data = uploaded_file.getvalue()
        image_parts = [{"mime_type": uploaded_file.type, "data": bytes_data}]
        return image_parts
    else:
        # Handle the error gracefully
        return []

# Main function
def main():
    st.header("AIHealth Ally 🤝: Nurturing Wellness Through Artificial Intelligence")

    # User input and file uploader
    custom_input_prompt = st.text_input("please give input on daily  activities, such as exercise routines, calorie intake, and water consumption.")

    # Set default prompt as an empty string
    default_input_prompt = default_prompt+" here is about my details of today which i follow daily : "+custom_input_prompt

    # Use default prompt if the custom prompt is explicitly left empty
    input_prompt = default_input_prompt

    uploaded_file = st.file_uploader("Choose an image of the invoice...", type=["jpg", "jpeg", "png"])

    # Display uploaded image
    if uploaded_file is not None:
        image = Image.open(uploaded_file)
        st.image(image, caption="Uploaded Image.", use_column_width=False)

    # Button to trigger processing
    if st.button("Delve into the Details"):
        # Process uploaded image details
        image_data = input_image_details(uploaded_file)

        # Get response from Gemini API
        response = say("Your prompt or guide goes here.", image_data, input_prompt)

        # Display the response
        st.subheader("Let's delve into the details of your meal 🍽️ ")
        st.write(response)

# Run the app
if __name__ == "__main__":
    main()
```

5Q Transfer learning is a powerful technique in machine learning where knowledge gained from one task is applied to enhance performance on a related but different task. This significantly improves efficiency and effectiveness, especially when data for the new task is limited.

Applications of Transfer Learning in Machine Learning Models:

Natural Language Processing (NLP): Pre-trained language models like BERT and GPT-3 are used for sentiment analysis, text summarization, machine translation, and more.
Computer Vision: Models trained on ImageNet, a massive dataset of labeled images, are fine-tuned for tasks like object detection, image classification, and medical image analysis.
Reinforcement Learning: Skills learned in simulated environments can be transferred to real-world robots for tasks like manipulation and navigation.
Anomaly Detection: Models trained on normal data can be used to detect abnormal patterns in new datasets, like identifying fraudulent transactions.
Beyond Machines: Can Humans Transfer Learn?

Absolutely! Transfer learning is not exclusive to machines. Humans constantly leverage previous experiences and knowledge to learn new things faster and more effectively.

Here are some examples:

Learning a new language: If you already know one language, it's easier to learn another with similar grammar or vocabulary.
Playing a new sport: Skills learned in one sport can translate to another, like hand-eye coordination from baseball helping with tennis.
Picking up a new instrument: Musicians can apply their musical knowledge to learn new instruments more quickly.
Job training: Previous work experience can be invaluable when transitioning to a new role with similar skills or knowledge requirements.
Key Differences:

While both machines and humans engage in transfer learning, there are some crucial differences:

Scale: Machines can process and analyze vast amounts of data, leading to more refined and complex knowledge transfer.
Generalization: Human transfer learning is often context-specific, while machines can sometimes generalize knowledge across broader domains.
Creativity: Humans excel at adapting and applying knowledge in novel ways, while machines usually require specific instructions for transfer.
In conclusion, transfer learning is a powerful tool utilized by both machines and humans to accelerate learning and improve performance. While their approaches differ, understanding these similarities and differences can inform the development of both artificial and human intelligence.
